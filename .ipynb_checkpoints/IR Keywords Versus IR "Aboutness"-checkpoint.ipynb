# Text Analysis: IR Keywords VS IR "Aboutness"

From Twitter to RefAnalytics, we've lately seen some interest in text analysis as a means of identifying trends and themes in loosely structured or unstructured data. Text corpora big and small excite curiosity and invite speculation, and from the standpoint of metadata one question is whether or how well descriptions of documents correlate with the received meaning or "aboutness" of the documents themselves. This is a topic we may be able to explore in our own insitutional repository, LoboVault. Specifically, we can compare the keywords assigned to documents within a collection to a word count and phrase analysis of those same documents.


## Thinking Algorithmically: The Big Picture

As an overview of the process, we want to:

1. Identify a collection of interest
2. Compile the subject metadata or keywords used in the collection
3. Compile the text corpus of documents from the collection
4. Analyze the corpus to derive single word counts, as well as 2- and 3- word phrased (ngrams)
5. Compare the assigned metadata against the output of step 4

## Get Document Text

```python
# Items in DSpace have internal IDs that are needed in order to access document properties via the API.

# 1. Get the Document ID (requires DSpace login): 28071

# 2. Create the API request

# Import necessary libraries - in this case, 'requests,' which provides a simple interface for making HTTP requests.
# Also, import 'string' for basic string functions.

import requests
import string

# Assign a variable with the base URL of the DSpace API. Any time we need this URL, we can reference it using the
# variable name.

baseURL = 'https://repository.unm.edu/rest'

# Referring to DSpace API documentation, the API request will also require a type and an item ID. For demonstration
# purposes, we can also assign these to variables.

entityType = 'items'
entityID = '28071'

# Referring to the requests documentation (http://docs.python-requests.org/en/master/), we request the document info
# via the DSpace API. The info we want relates to the item bitstreams - we're looking for the extracted PDF text.

# Note the API returns json data.

r = requests.get(baseURL + '/' + entityType + '/' + entityID + '?expand=bitstreams')

# Uncomment the next line to see the response.

# r.json()
```

```python
# We could parse the JSON programmatically to get the info we need, but for demo purposes
# what we want is the URL of the extracted text, given in the output above as '/bitstreams/114597/retrieve'.

# Making a slight modification to our HTTP request above, we can now request the document text.
# Note that we are reassigning the 'r' variable. 

# 3. Havest document text.

bitURL = '/bitstreams/114597/retrieve'

r = requests.get(baseURL + bitURL)

# Assign the response to a variable.

corpus = r.text

# Uncomment the next line to see the response
# print corpus
```

## Analyze Document Text

The code which follows has been adapted from two sources:

_Collocations_, accessed June 8, 2017. http://www.nltk.org/howto/collocations.html

_Word Cloud of the Most Frequent Words in the Canon of Sherlock Holmes_, accessed June 8, 2017. http://ramiro.org/notebook/sherlock-holmes-canon-wordcloud/.


```python
# NLTK refers to 'natural language toolkit,' and comes with many text analysis utilities.
# In addition to the library itself, we need to download some additional packages including the stopword list.

import nltk
# nltk.download('all')
```

```python
# Import additional libraries - from here on we are reusing old code!

# 1.1 Load stopwords (with libraries).

from nltk.collocations import *
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from scipy.misc import imread
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
from collections import Counter
```

```python
# Generate a list of most common words in the document text.
# First, "tokenize" the corpus into a list of words while removing whitespace characters.
# Second, use the 'most_common' method from collections library's Counter module to generate a list of
# most common terms in the corpus. Note that stopwords are excluded.

# 2. Load text to be analyzed and remove whitespace.

tokenizer = RegexpTokenizer(r'\w+')
cDataNoWS = tokenizer.tokenize(corpus)

# 3. Extract word count.

word_count_dict = Counter(w.title() for w in cDataNoWS if w.lower() not in stopwords.words())
word_count_dict.most_common()

# The output maps pretty well to the assigned keywords ('data curation', 'performing arts', and 'research methods'):
```

```python
# For fun we can also generate a word cloud:

import matplotlib as mpl
import matplotlib.pyplot as plt
import sys
from scipy.misc import imread
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
from wordcloud import WordCloud, STOPWORDS

limit = 50
infosize = 12

wordcloud = WordCloud(
    max_words=limit,
    background_color='#ffffff',
    mask=imread('cloud_mask_1.png')
).generate(', '.join(cDataNoWS))

fig = plt.figure()
fig.set_figwidth(9)
fig.set_figheight(12)

plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

### Bigram Analysis

A bigram is a two-word phrase. Below, we analyze the tokenized corpus data for bigrams and filter out stopwords. The stopword list is whatever is provided by default by NLTK.

Results are further filtered to limit the list of bigrams to those which occur three or more times.

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(cDataNoWS)
finder.apply_freq_filter(3)
ignored_words = nltk.corpus.stopwords.words('english')
finder.apply_word_filter(lambda w: len(w) < 3 or w.lower() in ignored_words)
finder.nbest(bigram_measures.likelihood_ratio, 100)

# Again, we see a pretty good correspondence with the assigned keywords.
```

### Trigram Analysis

A trigram is a three-word phrase. Below, we analyze the tokenized corpus data for trigrams and filter out stopwords. The stopword list is whatever is provided by default by NLTK.

Results are further filtered to limit the list of trigrams to those which occur three or more times.

```python
trigram_measures = nltk.collocations.TrigramAssocMeasures()
triFinder = TrigramCollocationFinder.from_words(cDataNoWS)
triFinder.apply_freq_filter(3)
ignored_words = nltk.corpus.stopwords.words('english')
triFinder.apply_word_filter(lambda w: len(w) < 3 or w.lower() in ignored_words)
triFinder.nbest(trigram_measures.likelihood_ratio, 50)

# This is harder to tell, but 'performing arts' and 'research methods' still stand out.
```

## Discussion

Based on the result, analysis of a larger dataset may be worthwhile. In order to scale up this process, some changes would be needed.

First, it's not practical to manually get all of the keywords assigned to all the documents within a collection. We would need to use an **iterator** to crawl the API document by document in order to harvest a **dictionary** of assigned subject terms and counts of those subject terms.

We would also need to iterate through the collection, again via the API, to harvest the extracted text for each document, which would also most likely be saved to a dictionary. We would likely bump up against memory constraints with a single dictionary containing all document text, so further refinements would be required to streamline the text processing. For example, we might not "save" the extracted text at all, but analyze each document on the fly and only compile the output of the word counts and bi/trigram analyses. Iterating through the collection via API will also require some JSON handling.
